```python
df, TBAs, params, all_coupons, rate_Base_raw, FN_issue, rate_Base = dl.load_all_data(data_folder = '../data/')
collapse_without_order, collapse_with_order = get_coupon_periods(df, all_coupons, FN_issue)
```

```python
p_45 =  pd.DataFrame(TBAs[4.5]['price']["2012-01-01":])
p_4 = pd.DataFrame(TBAs[4]['price']["2012-01-01":])
p_45 = p_45.merge(pd.DataFrame(p_45.pct_change()),how = "inner",left_index=True, right_index=True)
p_45 = p_45.rename(index=str, columns={"price_x": "price_45", "price_y": "return_45"})
p_4 = p_4.merge(pd.DataFrame(p_4.pct_change()),how = "inner",left_index=True, right_index=True)
p_4 = p_4.rename(index=str, columns={"price_x": "price_4", "price_y": "return_4"})
data = pd.merge(p_45,p_4,how = "inner",left_index=True, right_index=True)
data.dropna(inplace = True)
```

```python
data.head()
```

**Find Heding Duration**

```python
obs_mat = np.matrix(data["return_4"].values).T
kf_hr = KalmanFilter(transition_matrices=[1], observation_matrices=np.expand_dims(obs_mat, axis=1),
                          initial_state_mean = [0.6], initial_state_covariance = [1],
                          observation_covariance=[1], transition_covariance=[5])
state_means,state_covs = kf_hr.filter(np.matrix(data['return_45'].values).T)
data['hedge_ratio'] = state_means
```

```python
plt.plot(data['return_4'][:100].values)
plt.plot(data['return_45'][:100].values)
```

```python
plt.plot(data['return_4'][1200:].values)
plt.plot(data['return_45'][1200:].values)
```

```python
plt.figure(figsize=(10,6))
plt.rcParams['xtick.labelsize']=12
plt.rcParams['ytick.labelsize']=12
plt.plot(pd.to_datetime(data.index),data['hedge_ratio'].values,label='hedging ratio')
plt.xlabel('Time',fontsize=12)
plt.ylabel('Hedging Ratio',fontsize=12)
plt.title('Hedge Ratio between TBAs coupon 4.5 and TBAs coupon 4',fontsize=12)
#plt.plot(pd.to_datetime(data.index),data['hedge_ratio'].values)
```

```python
data.head()
```

```python
data["spread"] = data["return_45"] - data["hedge_ratio"]*data["return_4"] 
```

```python
plt.plot(pd.to_datetime(data.index),data["spread"].values)
```

```python
kf_s = KalmanFilter(transition_matrices=[1], observation_matrices=[1],
                          initial_state_mean = 0, initial_state_covariance = 1,
                          observation_covariance=1, transition_covariance=0.005)
state_means,state_covs = kf_s.filter(data['spread'].values)
data['filtered'] = state_means
```

```python
data.head()
```

```python
plt.figure(figsize=(10,6))
plt.rcParams['xtick.labelsize']=12
plt.rcParams['ytick.labelsize']=12
plt.plot(pd.to_datetime(data.index),data["spread"].values.cumsum(),label='Hedged Return',alpha=0.9)
plt.plot(pd.to_datetime(data.index),data["filtered"].values.cumsum(),label='Filtered Return',color = "gold")
plt.legend()
plt.xlabel('Time',fontsize=12)
plt.ylabel('Return',fontsize=12)
plt.title('Hedge Return and Filtered Return form KF',fontsize=12)
```

```python
data['alpha'] = data['spread'] - data['filtered']
data['position'] = -(data['alpha'] - data['alpha'].rolling(60).mean())/data['alpha'].rolling(60).std()
data_clean = data.dropna(axis=0)
```
